```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.datasets import load_iris
iris = load_iris()
```

```python
X = iris.data[:, 2] # 입력값
y = iris.data[:, 3] # 목표값

X = iris.data[:, [2]]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression() # Regression --> 회귀
model.fit(X, y) # 훈련
y_pred = model.predict(X) # X로 y예측, predict --> 예측하다
print(y_pred)
```

```python
plt.scatter(X[:, 0], y, c=iris.target) # 원본 데이터
plt.plot(X[:, 0], y_pred, 'r') # 예측 데이터
plt.colorbar()
plt.axis('scaled') # 단위를 같게 한다..
```

```python
model.score(X, y)
```

```python
model.coef_ # 가중치?
```

```python
model.intercept_
```

```python
X = iris.data[:, :2]
y = np.array([0] * 50 + [1] * 100) # 0 50개, 1 100개 생성

X.shape, y.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X, y)
y_pred = model.predict(X)
plt.subplot(1, 1, 1)
plt.imshow(y_pred.reshape(10,15))
```

```python
plt.imshow(y.reshape(10, 15))
```

```python
w = model.coef_[0]
b = model.intercept_[0]

w, b
```

```python
xs = np.array([4, 8])
ys = -(w[0]*xs + b) / w[1]

plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.5)
plt.plot(xs, ys, 'r')
plt.axis('scaled')
```

```python
X = iris.data[50:, [0, 1]]
y = np.array([0]*50 + [1] * 50)
```

```python
model = LogisticRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
w = model.coef_[0]
b = model.intercept_[0]

w, b
```

```python
xs = np.array([4, 8])
ys = -(w[0]*xs + b)/w[1]

plt.scatter(X[:,0], X[:,1], c=y)
plt.colorbar()
plt.plot(xs, ys, 'r')
plt.axis('scaled')
plt.ylim([1.5, 4.5])
```

```python
import mglearn
```

```python
mglearn.plots.plot_2d_separator(model, X, fill=True, eps=0.5, alpha=0.4)
mglearn.discrete_scatter(X[:,0], X[:,1], y)
plt.legend(loc=4)
```
